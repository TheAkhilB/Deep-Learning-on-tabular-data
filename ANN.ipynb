```python
import pandas as pd
```

```python
df = pd.read_csv("index.csv")
```

```python
df.head()
```

```python
X = df.drop(columns = ["Creditability"])
y = df['Creditability']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

```python
y_train = y_train.to_numpy()
```

```python
y_train
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.optimizers import Adamax, Adam, RMSprop, SGD
```

```python
# Dense -> adding Hidden input layers
ann_variable = tf.keras.models.Sequential()

ann_variable.add(tf.keras.layers.Dense(units=8,activation='relu'))

ann_variable.add(tf.keras.layers.Dense(units=8,activation='relu'))

ann_variable.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

ann_variable.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann_variable.fit(X_train,y_train,batch_size=32,epochs=50)
```

```python
Y_pred = ann_variable.predict(X_test)
Y_pred
```

```python
Y_pred = (Y_pred > 0.5)
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
confusion_matrix(Y_pred,y_test)
```

```python
cm_ANN = confusion_matrix(y_test, Y_pred)
```

```python
tn, fp, fn, tp = cm_ANN.ravel()
```

```python
precision_ann = precision_score(y_test, Y_pred)
print("precision = "+str(precision_ann))
recall_ann = recall_score(y_test, Y_pred)
print("recall = "+ str(recall_ann))
accuracy_ann = accuracy_score(y_test, Y_pred)
print("accuracy = "+ str(accuracy_ann))
F1_score_ann = f1_score(y_test, Y_pred)
print("F1 Score = "+ str(F1_score_ann))
```

```python
specificity = (tn)/(tn+fp)
print("specificity ="+ str(specificity))
```
